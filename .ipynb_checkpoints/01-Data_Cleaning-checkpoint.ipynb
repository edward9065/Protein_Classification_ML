```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pdb_metadata = pd.read_csv("pdb_data_no_dups.csv")
```

```python
pdb_sequence_data = pd.read_csv("pdb_data_seq.csv")
```

```python
pdb_metadata = pdb_metadata.loc[pdb_metadata["macromoleculeType"] == "Protein"]
```

```python
pdb_sequence_data = pdb_sequence_data[pdb_sequence_data["macromoleculeType"] == "Protein"]
```

```python
structure_id_with_NA = pdb_sequence_data[pdb_sequence_data.isna().any(axis=1)]["structureId"]
pdb_sequence_data_filtered = pdb_sequence_data.loc[~pdb_sequence_data["structureId"].isin(structure_id_with_NA)]
```

```python
# Sort by structure_id and chain_id to ensure alphabetical order of chains
df_sorted = pdb_sequence_data_filtered.sort_values(by=['structureId', 'chainId'])

# Group by structure_id and concatenate sequences in alphabetical order of chain_id
df_grouped = df_sorted.groupby('structureId')['sequence'].apply('+'.join).reset_index()

# Rename the resulting column to indicate concatenated sequences
df_grouped.columns = ['structureId', 'concatenated_sequence']

df_grouped
```

```python
pdb_metadata_filtered = pdb_metadata.dropna(subset=['classification'])
```

```python
pdb_metadata_filtered
```

```python
protein_sequences_labelled = merged_df = pd.merge(df_grouped, pdb_metadata_filtered[["structureId", "classification"]], on='structureId', how='inner')  
```

```python
protein_sequences_labelled
```

```python
lengths = protein_sequences_labelled['concatenated_sequence'].apply(len)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(lengths, bins=30, color='blue', alpha=0.7, edgecolor='black')  # Adjust bins as needed

# Set logarithmic scale for the y-axis
plt.yscale('log')

plt.title('Histogram of Protein Sequence Lengths (Log Scale)')
plt.xlabel('Sequence Length')
plt.ylabel('Frequency (Log Scale)')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
threshold = 100000 # Set your desired threshold

# Remove rows where the length of the 'concatenated_sequence' exceeds the threshold
protein_sequences_labelled = protein_sequences_labelled[
    protein_sequences_labelled['concatenated_sequence'].apply(len) <= threshold
]

# Optional: Check the shape of the DataFrame after filtering
print(f"Remaining rows: {protein_sequences_labelled.shape[0]}")
```

```python
protein_sequences_labelled
```

```python
protein_sequences_labelled['classification'] = protein_sequences_labelled['classification'].str.upper()
```

```python
classification_counts = protein_sequences_labelled['classification'].value_counts()
min_samples_threshold = 500
valid_classifications = classification_counts[classification_counts >= min_samples_threshold].index
filtered_protein_sequences = protein_sequences_labelled[protein_sequences_labelled['classification'].isin(valid_classifications)]
```

```python
filtered_protein_sequences.to_csv("filtered_labelled_protein_sequences.tsv", sep="\t")
```
